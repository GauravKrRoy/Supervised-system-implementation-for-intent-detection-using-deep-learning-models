```python
data = raw_train + raw_test
```

```python
idata = raw_itrain + raw_itest
```

```python
word_dict = dict()
i_w = 1

for utterance in data:
    words = utterance.strip().split()
    for word in words:
        if(word not in word_dict):
            word_dict[word] = i_w
            i_w = i_w + 1
```

```python
intent_dict = dict()
i_i = 0

for utterance in idata:
    intents = utterance.strip().split()
    for intent in intents:
        if(intent not in intent_dict):
            intent_dict[intent] = i_i
            i_i += 1
```

```python
pos_dict = dict()
i = 0

for utterance in data:
    tags = nltk.pos_tag(utterance.split())
    for tag in tags:
        tag = tag[1]
        if(tag not in pos_dict):
            pos_dict[tag] = i
            i += 1
```

# One Hot Representations

```python
n = len(intent_dict)

y_train_intent = list()
for intent in itrain:
    out = np.zeros(n)
    out = list(out)
    out[intent_dict[intent]] = 1
    y_train_intent.append(out)
    
y_test_intent = list()
for intent in itest:
    out = np.zeros(n)
    out = list(out)
    out[intent_dict[intent]] = 1
    y_test_intent.append(out) 
```
